```python
def my_function2(x, y, z=1.5):
    """Evaluate a condition and based on it execute a specific function """
    if z > 1:
        return z * (x + y)
    else:
        return z / (x + y)
```

```python
my_function2(5, 6, z=0.7)
```

```python
my_function2(3.14, 7, 3.5)
```

```python
my_function2(10,20)
```
